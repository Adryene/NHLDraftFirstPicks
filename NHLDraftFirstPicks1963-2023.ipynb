# NHL Top Draft Picks Analysis (1963-2023)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reads the Top Draft Picks over NHL History
file_path = 'NHLTopDraftPicks1963-2023.csv' 
df = pd.read_csv(file_path)
df
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df
```

```python
# Shows all Column names and their Data Types
df.info()
```

```python
# Shows general information about number values
df.describe()
```

# First Pick Position Statistics

```python
# Sort by Position to find percentages of First Picks by position
position_count = df['Position'].value_counts()
position_percentages = position_count / position_count.sum() * 100

# Prints the data above graph 
print(position_percentages.to_string(index=True))

# Create pie chart to show distribution of player Positions
plt.figure(figsize=(12, 8))
plt.pie(position_percentages, labels=position_percentages.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('First Pick Position Distribution')
plt.show()
```

# First Pick Country of Origin Statistics

```python
# Sort by Country to find percentages of First Pick player Country of Origin
country_count = df['Country'].value_counts()
country_percentages = country_count / country_count.sum() * 100

# Prints the data above graph 
print(country_percentages.to_string(index=True))

# Create pie chart to show distribution of players' Country of Origin
plt.figure(figsize=(12, 8))
plt.pie(country_percentages, labels=country_percentages.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('First Pick Country of Origin Distribution')
plt.show()
```

# First Pick Team Distribution

```python
# Sort by Team to find percentages of which teams get First Pick 
team_count = df['Team'].value_counts()
team_percentages = team_count / team_count.sum() * 100

# Sort the percentages in descending order and select the top 5
top_5_teams = team_percentages.nlargest(5)

# Prints Top 5 Teams with highest First Pick percentages  
print(top_5_teams.to_string(index=True))

# Create pie chart to show distribution of Team First Picks
plt.figure(figsize=(12, 8))
plt.pie(team_percentages, labels=team_percentages.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('First Pick Team Distribution')
plt.show()
```

# Tallest First Pick Player Statistics

```python
# Sort the dataframe by height in descending order to find the tallest players
df.sort_values(by="Height (Inches)", ascending=False).head()
```

```python
# Sort the df by height in descending order to find the tallest players
sorted_df = df.sort_values(by="Height (Inches)", ascending=False).head()

# Print the data above the graph
print(sorted_df[["Player","Height (Inches)"]].to_string(index=False))

# Creates a vertical bar chart
plt.figure(figsize=(12, 8))
plt.bar(sorted_df["Player"], sorted_df["Height (Inches)"])
plt.xlabel("Player")
plt.ylabel("Height (Inches)")
plt.title("Tallest NHL Draft First Picks")
plt.show()
```
